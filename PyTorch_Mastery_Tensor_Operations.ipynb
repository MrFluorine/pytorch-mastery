# Tensor Operations

```python
import torch
```

```python
# Tensor = Menory + shape + stride
# Most operations like view(), transpose(), permute(), squeeze(), unsqueeze() and expand() only change SHAPE and STRIDE not memory
```

```python
# ============================================================
# 1. SHAPE, RANK, DTYPE, DEVICE
# ============================================================
x = torch.randn(3, 4, 5)
print("Shape:", x.shape)
print("Rank :", x.ndim)
print("Dtype:", x.dtype)
print("Device:", x.device)
# Total elements
print("Elements:", x.numel())
```

```python
# ============================================================
# 2. STRIDES
# ============================================================
x = torch.arange(60).view(3,4,5)
print(x)
print("Shape:", x.shape)
print("Strides:", x.stride())
y = x.reshape(2, 6, 5)
print(y)
print("Shape:", y.shape)
print("Strides:", y.stride())
```

```python
# ============================================================
# 3. CONTIGUOUS TENSORS - If I walk through the tensor logically, do I also walk through memory sequentially?
# ============================================================
x = torch.arange(60).view(3,4,5)
print(x)
print("CONTIGUOUS TENSORS?:", x.is_contiguous())
y = x.transpose(0,2)
print(y)
print("CONTIGUOUS TENSORS?:", y.is_contiguous())
```

```python
# ============================================================
# 4. VIEWS SHARE MEMORY
# ============================================================
x = torch.arange(60)
print(x)
y = x.view(3,4,5)
print(y)
y[0,0,0] = 99
print(x)
# x changes because:
# x and y point to same memory
```

```python
# ============================================================
# 5. CLONE
# ============================================================
x = torch.arange(12)
print(x)
y = x.clone()
y[0] = 99
print(x)

# clone() creates new memory
```

```python
# ============================================================
# 6. RESHAPE - It may create new memory but is contiguous
# ============================================================
x = torch.arange(60)
y = x.reshape(2,6,5)
print(y.shape)
print(y.is_contiguous())
```

```python
# ============================================================
# 7. TRANSPOSE - Makes the tensor non contagiuous
# ============================================================
x = torch.arange(12).view(3,4)
print(x.shape)
y = x.transpose(0,1)
print(y.shape)
print(y.is_contiguous())
y[0,0] = 99
print(x)
y = y.contiguous()
print(y.is_contiguous())
y[0,0] = 0
print(x)
```

```python
# ============================================================
# 8. VIEW ERROR -  View() only works
# ============================================================
x = torch.arange(60).reshape(3, 4, 5)
y = x.transpose(0,2)
try:
    z = y.view(60)
except Exception as e:
  print(e)
# Fix:
z = y.contiguous().view(60)
print(z)
```

```python
# ============================================================
# 9. PERMUTE
# ============================================================
x = torch.randn(2,3,4,5)
print(x.shape)
y = x.permute(2,0,3,1)
print(y.shape)
```

```python
# ============================================================
# 10. UNSQUEEZE
# ============================================================
x = torch.arange(12)
print(x)
print(x.shape)
y = x.unsqueeze(0)
print(y.shape)
print(y)
y = x.unsqueeze(1)
print(y.shape)
print(y)
```

```python
# ============================================================
# 11. SQUEEZE
# ============================================================
x = torch.randn(1,3,1,4,1)
print(x.shape)
y = x.squeeze()
print(y.shape)
```

```python
# ============================================================
# 12. BROADCASTING
# ============================================================
A = torch.randn(3,4)
B = torch.randn(4)
C = A + B
print(C.shape)

A = torch.randn(2,4,5)
B = torch.randn(1,4,1)
C = A + B
print(C.shape)
```

```python
# ============================================================
# 13. EXPAND
# ============================================================
x = torch.tensor([[1,2,3]])
print(x.shape)
y = x.expand(5,3)
print(y.shape)

# No new memory
# Uses broadcasting tricks

```

```python
# ============================================================
# 14. REPEAT
# ============================================================

x = torch.tensor([[1,2,3]])
y = x.repeat(5,2)
print(y.shape)
# New memory allocated
print(y)
```

```python
# ============================================================
# 15. MATRIX MULTIPLICATIONS
# ============================================================
A = torch.randn(2, 3, 4)
B = torch.randn(4, 5)
C = A@B
print(C.shape)
x = torch.randn(2, 4, 5)
D = torch.matmul(A,x)
print(D.shape)
```

```python
# ============================================================
# 16. SOFTMAX
# ============================================================
x = torch.randn(3,4)
print(x)
y = torch.softmax(x, dim=1)
print(y)
```

```python
import torch

# ============================================================
# 1. MATRIX MULTIPLICATION - using for loops
# ============================================================
A = torch.randn(2, 3, 4)
B = torch.randn(4, 5)
def matrixmul(A: torch.Tensor, B: torch.Tensor) -> torch.Tensor:
    T_shape = list(A.shape)
    T_shape[-1] = B.shape[-1]
    T = torch.zeros(T_shape)
    if A.shape[-1] != B.shape[-2]:
        raise ValueError("A.shape[-1] != B.shape[-2]")
    else:
      if len(A.shape) == 2:
        for i in range(A.shape[0]):
          for j in range(B.shape[1]):
            for k in range(B.shape[0]):
              T[i,j] += A[i,k] * B[k,j]
      elif len(A.shape) == 3:
        for b in range(A.shape[0]):
          for i in range(A.shape[1]):
            for j in range(B.shape[1]):
              for k in range(B.shape[0]):
                T[b,i,j] += A[b,i,k] * B[k,j]
      else:
        raise ValueError("A.shape[-1] != B.shape[-2]")
    return T


C = matrixmul(A,B)
c_torch = A@B
print(torch.allclose(C, c_torch))
```

```python
# ============================================================
# 2. MATRIX MULTIPLICATION - using einsum
# ============================================================

A = torch.randn(2, 3, 4)
B = torch.randn(4, 5)
def matrixmul(A: torch.Tensor, B: torch.Tensor) -> torch.Tensor:
    if A.shape[-1] != B.shape[-2]:
        raise ValueError("A.shape[-1] != B.shape[-2]")
    else:
      if len(A.shape) == 2:
        T = torch.einsum('ij,jk->ik', A, B)
      elif len(A.shape) == 3 and len(B.shape) == 3:
        T = torch.einsum('bij,bjk->bik', A, B)
      elif len(A.shape) == 3 and len(B.shape) == 2:
        T = torch.einsum('bij,jk->bik', A, B)
      else:
        raise ValueError("A.shape[-1] != B.shape[-2]")
    return T



C = matrixmul(A,B)
c_torch = A@B
print(torch.allclose(C, c_torch))
```

```python
# ============================================================
# 2. SOFTMAX
# ============================================================
def denji_softmax(A: torch.Tensor, dim: int) -> torch.Tensor:
    if dim+1 > len(A.shape):
      raise ValueError("Index out of range")
    if dim < 0:
      dim += len(A.shape)

    x = A.transpose(dim, -1)
    original_shape = x.shape
    x = x.reshape(-1, x.shape[-1])
    output = torch.zeros_like(x)
    for row in range(x.shape[0]):
      max_val = x[row].max()
      output[row] = torch.exp(x[row] - max_val)
      output[row] = output[row] / output[row].sum()
    output = output.reshape(original_shape)
    output = output.transpose(dim, -1)
    return output




```
